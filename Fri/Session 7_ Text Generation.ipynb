```python
from transformers import AutoModelForCausalLM, AutoTokenizer,T5ForConditionalGeneration
import torch
```

```python
torch_device = "cuda" if torch.cuda.is_available() else "cpu"
# connect to GPU runtime to check this
print(f'(Free memory, Available Memory){torch.cuda.mem_get_info()}')
```

```python
prompt1 = "I went to the bakery to"
prompt2 = "negate: I went to the bakery"
prompt3 = "summarize: I went to the bakery and bought a baguette and some muffins. Then I went to the Bank and deposited some money. Then I went for Lunch with a friend."
```

# Decoder Only Models

### GPT

```python
tokenizer_gpt = AutoTokenizer.from_pretrained("openai-gpt")
model_gpt = AutoModelForCausalLM.from_pretrained("openai-gpt", pad_token_id=tokenizer_gpt.eos_token_id).to(torch_device) # add the EOS token as PAD token to avoid warnings
```

```python
input_ids_gpt = tokenizer_gpt.encode(prompt1, return_tensors="pt").to(torch_device)
output_gpt = model_gpt.generate(input_ids_gpt, max_length=40, pad_token_id=tokenizer_gpt.eos_token_id).to(torch_device)
print(tokenizer_gpt.decode(output_gpt[0], skip_special_tokens=True))
```

```python
input_ids_gpt = tokenizer_gpt.encode(prompt2, return_tensors="pt").to(torch_device)
output_gpt = model_gpt.generate(input_ids_gpt, max_length=40, pad_token_id=tokenizer_gpt.eos_token_id).to(torch_device)
print(tokenizer_gpt.decode(output_gpt[0], skip_special_tokens=True))
```

```python
input_ids_gpt = tokenizer_gpt.encode(prompt3, return_tensors="pt").to(torch_device)
output_gpt = model_gpt.generate(input_ids_gpt, max_length=40, pad_token_id=tokenizer_gpt.eos_token_id).to(torch_device)
print(tokenizer_gpt.decode(output_gpt[0], skip_special_tokens=True))
```

CTRL via Huggingface API

```python
import requests
```

```python
API_TOKEN = "ADD YOUR TOKEN" # Please do not use my API token (shown here for demonstration purposes), use yours only.
headers = {"Authorization": f"Bearer {API_TOKEN}"}
API_URL = "https://api-inference.huggingface.co/models/Salesforce/ctrl"
def query(payload):
    response = requests.post(API_URL, headers=headers, json=payload)
    return response.json()
```

```python
ctrl_style = "Conspiracy "
data = query({"inputs": ctrl_style + prompt1, "wait_for_model": True})
data
```
